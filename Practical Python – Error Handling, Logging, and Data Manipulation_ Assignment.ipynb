Q1 . What is the difference between multithreading and multiprocessing?

Ans. Multithreading

1. Multiple threads of execution within a single process.
2. Threads share the same memory space.

Ans. Multiprocessing

1. Multiple independent processes running concurrently.
2. Each process has its own separate memory space

Q4 Write a Python program that reads from one file and writes its content to
another file.

```python
with open('input.txt', 'r') as source_file:
    content = source_file.read()

with open('output.txt', 'w') as destination_file:
    destination_file.write(content)

print("Content copied successfully!")
```

Q5 Write a program that handles both IndexError and KeyError using a
try-except block.


```python
list1 = [10, 20, 30]
my_dict = {"name": "Alice", "age": 25}

try:
      print("List item:", list1[3])

      print("Dictionary value:", my_dict["address"])

except IndexError:
    print("Oops! You tried to access an index that doesn't exist in the list.")

except KeyError:
    print("Oops! That key is not found in the dictionary.")

finally:
    print("Program finished.")
```

Q10 Calculate the correlation matrix using Seaborn and visualize it with a
heatmap.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100) * 2,
    'C': np.random.randn(100) + 1,
    'D': np.random.randn(100) * 0.5 - 1})


corr_matrix = data.corr()

# Visualize with Seaborn heatmap
sns.heatmap(corr_matrix, annot=True, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
